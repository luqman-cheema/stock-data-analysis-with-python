# Problem - 4: Use Webscraping to Extract GME Revenue Data



This Python script demonstrates how to extract and process financial data for GameStop (GME) using web scraping. It begins by installing necessary libraries like bs4 for BeautifulSoup and suppressing warnings to maintain code clarity. The HTML content of the webpage containing GameStop's revenue data is retrieved and parsed to find the relevant table. The table is transformed into a pandas dataframe, renamed with appropriate column headers, and structured for analysis. This approach showcases efficient data extraction and preparation techniques for financial analytics.

```python
!pip install bs4
!pip install nbformat

import pandas as pd
import requests
from bs4 import BeautifulSoup

import warnings
# Ignore all warnings
warnings.filterwarnings("ignore", category=FutureWarning)

gme_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/stock.html"
html_data2 = requests.get(gme_url).text

gme_soup = BeautifulSoup(html_data2,"html.parser")

table = gme_soup.find("table")
gme_revenue = pd.read_html(str(table))[0]
gme_revenue.columns = ['Date', 'Revenue']

gme_revenue.tail()
```
