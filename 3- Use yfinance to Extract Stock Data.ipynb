# Problem - 3: Use yfinance to Extract Stock Data



This script highlights how data analysts can efficiently use the yfinance library to gather historical stock data. It focuses on GameStop (GME), creating a ticker object to retrieve maximum available historical price data. To ensure seamless execution, warnings are suppressed for clarity. The data is structured into a pandas dataframe by resetting the index, enabling straightforward analysis. This concise approach is a must-have for financial analysts exploring stock trends and market insights.

```python
!pip install yfinance

import yfinance as yf
import warnings
# Ignore all warnings
warnings.filterwarnings("ignore", category=FutureWarning)

game_stock_ticker = yf.Ticker("GME")

gme_data = game_stock_ticker.history(period="max")

gme_data.reset_index(inplace=True)
gme_data.head()
```
