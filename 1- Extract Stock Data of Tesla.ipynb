# Problem - 1: Extract Stock Data of Tesla



This Python script showcases how data analysts can utilize the yfinance library to extract comprehensive stock market data. By creating a ticker object for Tesla (ticker symbol: TSLA), it retrieves extensive historical stock price data. The dataset is then reset to a structured dataframe format for further analysis. The .info method provides additional metadata about the stock, while the preview of the dataframe allows analysts to inspect the extracted data quickly. This code is an essential tool for analyzing market trends and gaining insights into stock performance.

```python
!pip install yfinance

import yfinance as yf

tesla_ticker = yf.Ticker("TSLA")
tesla_ticker.info

tesla_data = tesla_ticker.history(period="max")

tesla_data.reset_index(inplace=True)
tesla_data.head()

```
